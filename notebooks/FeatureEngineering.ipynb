```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder, StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pickle as pkl

%matplotlib inline
```

```python
df = pd.read_csv(r"..\data\heart_cleaned.csv")
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)

df.shape
```

```python
df.columns
```

### Ordinal Encoding

```python
categories_order = [['Zero', 'One', 'Two', 'Three', 'Four']]
orde = OrdinalEncoder(categories=categories_order)
```

```python
df['vessels_colored_by_flourosopy_encoded'] = orde.fit_transform(df[['vessels_colored_by_flourosopy']])
```

```python
df['vessels_colored_by_flourosopy_encoded'].value_counts()
```

```python
categories_order_slope = [['Upsloping', 'Flat', 'Downsloping']]
ordencode = OrdinalEncoder(categories=categories_order_slope)

df[['slope_encode']] = ordencode.fit_transform(df[['slope']])
```

```python
print(df['slope_encode'].value_counts())
print(df['slope'].value_counts())
```

### Label Encoding

```python
label_enc = LabelEncoder()
```

```python
df['sex_enc'] = label_enc.fit_transform(df['sex'])
```

```python
print(df['sex'].value_counts())
print(df['sex_enc'].value_counts())
```

```python
df['fasting_blood_sugar_encode'] = label_enc.fit_transform(df['fasting_blood_sugar'])
```

```python
print(df['fasting_blood_sugar'].value_counts())
print(df['fasting_blood_sugar_encode'].value_counts())
```

```python
df['exercise_induced_angina_encode'] = label_enc.fit_transform(df['exercise_induced_angina'])
```

```python
print(df['exercise_induced_angina'].value_counts())
print(df['exercise_induced_angina_encode'].value_counts())
```

```python
df_encoded = df.copy()
df_encoded
```

```python
df_encoded = df_encoded.drop(columns=['vessels_colored_by_flourosopy', 'slope', 'sex', 'fasting_blood_sugar','exercise_induced_angina'])
df_encoded.head()
```

```python
df_encoded.rename(columns={'vessels_colored_by_flourosopy_encoded':'vessels_colored_by_flourosopy',
       'slope_encode':'slope', 'sex_enc':'sex', 'fasting_blood_sugar_encode':'fasting_blood_sugar',
       'exercise_induced_angina_encode':'exercise_induced_angina'}, inplace=True)
```

```python
df_encoded.to_csv('../data/heart_feature_encoded.csv', index=False)
```

### One-Hot Encoding

```python
one_hot = OneHotEncoder(sparse_output=False)
```

```python
cat_one_hot = ['chest_pain_type', 'rest_ecg', 'thalassemia']

one_hot_encoded = one_hot.fit_transform(df[cat_one_hot])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=one_hot.get_feature_names_out(cat_one_hot))
one_hot_df.head()
```

```python
df_encode = pd.concat([df, one_hot_df], axis=1)
```

```python
df_encode.head()
```

```python
cat_col = ['sex', 'chest_pain_type', 'fasting_blood_sugar',	'rest_ecg',	'exercise_induced_angina',	'slope', 'vessels_colored_by_flourosopy', 'thalassemia']
```

```python
df_encode = df_encode.drop(cat_col, axis=1)
df_encode.head()
```
